# 1. Import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2. Define a two-dimensional function: $f(x) = x^2 + 2 \cdot x $

```python
def myFunction(x):
    f = x**2 + 2*x
    return f
```

# 3. Define derivations of the given function: $f'(x) = 2 \cdot x + 2$, $f''(x) = 2$

```python
def Derivative(x):
    Deriv = 2*x + 2
    return Deriv
def DDerivative(x):
    DDeriv = 2
    return DDeriv
```

# 4. Define approximation functions of given function for three different points

```python
# Approximation = f(a) + f'(a)(x-a) + 1/2*f''(a)(x-a)^2 + ...
def Approx_2D_minus7(x):
    # myFunction(-7) + Derivative(-7)(x+7) + 1/2*2*(x+7)^2 = x^2 + 2x
    Approximation = x**2 + 2*x
    return Approximation
def Approx_2D_plus3(x):
    # myFunction(3) + Derivative(3)(x-3) + 1/2*2*(x-3)^2 = x^2 + 2x
    Approximation = x**2 + 2*x
    return Approximation
def Approx_2D_plus9(x):
    # myFunction(9) + Derivative(9)(x-9) + 1/2*2*(x-9)^2 = x^2 + 2x
    Approximation = x**2 + 2*x
    return Approximation
```

# 5. Define the domain of the given function: $x = [-10 : 0.1 : 10]$

```python
x = np.arange(-10, 10, 0.1)
```

# 6. Compute the graph

```python
f = myFunction(x)
Df = Derivative(x)
DDf = DDerivative(x)
fhat_1 = Approx_2D_minus7(x)
fhat_2 = Approx_2D_plus3(x)
fhat_3 = Approx_2D_plus3(x)
```

# 7. Plot graphs for the function, its derivative, and approximation results

```python
plt.figure(2)
plt.plot(x, f, 'b.', label="function")
plt.plot(x, Df, 'ro', label="derivative")

plt.plot(x, fhat_1, 'gv', label="function hat_1 (around a = -7)")
plt.plot(x, fhat_2, 'c^', label="function hat_2 (around a = 3)")
plt.plot(x, fhat_3, 'm*', label="function hat_3 (around a = 9)")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
